# Explorasi Data

## **Pokemon Mega Evolution**

```python
#Explorasi Data
#Mencari berapa banyak pokemon yang memiliki form Mega
mega_pokemon = df['Name'].str.contains('\s*Mega\s' , regex=True)
df_mega = df[mega_pokemon]
```

```python
df_mega
```

```python
len(df_mega.index)
```

```python
#Digunakan untuk mencari pokemon dengan form mega evolution yang memiliki nilai total tertinggi
df_mega.sort_values(by='Total',ascending=False).head(5)
```

## **Pokemon Legendary**

```python
df_legendary = df[df['Legendary'] == True]
```

```python
df_legendary
```

```python
legendary_mega = df_legendary['Name'].str.contains("\s*Mega\s", regex=True)
```

```python
#Drop Pokemon yang masih memiliki nama Mega
df_legendary = df_legendary[~legendary_mega]
```

```python
df_legendary
```

```python
legendary_primal = df_legendary['Name'].str.contains("\s*Primal\s", regex=True)
```

```python
#Drop Pokemon yang masih memiliki Name Primal
df_legendary = df_legendary[~legendary_primal]
```

```python
df_legendary
```

```python
df_legendary.sort_values(by='Total',ascending=False).head(5)
```

## **Pokemon Non-Legendary**

```python
#Membuat dataframe khusus pokemon non legendary
df_non_legendary = df[~df['Name'].str.contains('\s*Mega\s' , regex=True)]
```

```python
#Drop Legendary
df_non_legendary = df_non_legendary[df_non_legendary['Legendary'] == False]
```

```python
df_non_legendary
```

```python
df_non_legendary.sort_values(by='Total',ascending=False).head(5)
```
